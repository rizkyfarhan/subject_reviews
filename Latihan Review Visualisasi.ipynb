```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib._color_data as mcd
%matplotlib inline
```

```python
youtube = pd.read_csv('USvideos.csv')
```

```python
youtube.info()
```

```python
youtube.head()
```

```python
youtube.describe()
```

## Latihan Distribution Plot dengan Seaborn

```python
sns.reset_defaults()
sns.distplot(youtube['likes'])
```

```python
youtube[['likes','dislikes']].corr()
```

```python
sns.jointplot(x='likes',y='dislikes',data=youtube,kind='reg')
```

```python
g = sns.PairGrid(youtube[['category_id','views','likes','dislikes','comment_count']])
g.map_diag(sns.kdeplot)
g.map_upper(sns.regplot)
g.map_lower(sns.regplot)
```

## Categorical Plot

```python
sns.barplot(x='comments_disabled',y='likes',data=youtube)
```

```python
sns.countplot(youtube['comments_disabled'])
```

```python
sns.boxplot(x='likes',y='comments_disabled',data=youtube,orient='h')
```

```python
sns.violinplot(x='comments_disabled',y='likes',data=youtube)
sns.stripplot(x='comments_disabled',y='likes',data=youtube,jitter=True)
```
